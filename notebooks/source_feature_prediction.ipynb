```python
# ===== Standard Library =====
import os
import sys
import pandas as pd

# Add ../src to path
sys.path.append("../src")
ROOT = os.path.abspath(os.path.join(os.getcwd(), ".."))

from metadata.loading import load_datasets
from feature_prediction import build_dataset
```

```python
datasets = load_datasets()
X_heart, y_heart = build_dataset(datasets['mixed_data'], organ="heart")
X_lung, y_lung = build_dataset(datasets['mixed_data'], organ="lung")
```

```python
from feature_prediction.feature_regression import (
    train_rf_regressor,
    evaluate_rf_model
)

# Heart model
rf_heart = train_rf_regressor(X_heart, y_heart, target_name="heart")
results_heart, results_heart_df = evaluate_rf_model(rf_heart, X_heart, y_heart, n_boot=500, save_path="outputs/source_reconstruction/rf_heart_feature_predictions.csv")

# Lung model
rf_lung = train_rf_regressor(X_lung, y_lung, target_name="lung")
results_lung, results_lung_df = evaluate_rf_model(rf_lung, X_lung, y_lung, n_boot=500, save_path="outputs/source_reconstruction/rf_lung_feature_predictions.csv")

rf_results = pd.concat([results_heart_df, results_lung_df])
rf_results
```
